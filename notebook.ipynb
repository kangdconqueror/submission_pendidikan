## Data Preprocessing

### Seleksi Fitur

```python
import pandas as pd

# --- 1) Daftar fitur terpilih (hasil insight EDA) --------------------------
feat_keep = [
    # korelasi positif
    'Curricular_units_2nd_sem_approved',
    'Curricular_units_2nd_sem_grade',
    'Curricular_units_1st_sem_approved',
    'Curricular_units_1st_sem_grade',
    'Tuition_fees_up_to_date',
    'Scholarship_holder',

    # korelasi negatif
    'Debtor',
    'Age_at_enrollment',
    'Gender',
    'Application_mode',

    # kolom target
    'Status'
]

# --- 2) Sub‑set data‑frame ---------------------------------------------------
df_sel = df[feat_keep].copy()

# --- 3) Cek cepat hasilnya ---------------------------------------------------
print(df_sel.shape)       # (4424, 11)  → 10 fitur + 1 target
df_sel.head()
```

```python
df_sel.info()
```

### Memisahkan Fitur dan Target

Kolom target adalah Status (Dropout, Enrolled, Graduate), sedangkan yang lain adalah fitur.

```python
# target
y = df_sel['Status']          # Dropout / Enrolled / Graduate
X = df_sel.drop(columns='Status')
```

### Encoding Target (Label Encoding)

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_encoded = le.fit_transform(y)     # Simpan le untuk inverse_transform saat evaluasi

# Lihat hasil label
dict(zip(le.classes_, le.transform(le.classes_)))
```

### Split Data (Train-Test)

```python
from sklearn.model_selection import train_test_split

# stratify agar distribusi target seimbang
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)
```

## Modeling

### Train Model Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
rf_preds = rf.predict(X_test)
```

### Train Model Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)
lr_preds = lr.predict(X_test)
```

### Evaluasi Random Forest

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print("Accuracy:", accuracy_score(y_test, rf_preds))
print(confusion_matrix(y_test, rf_preds))
print(classification_report(y_test, rf_preds, target_names=le.classes_))
```

### Evaluasi Logistic Regression

```python
print("\n=== Logistic Regression ===")
print("Accuracy:", accuracy_score(y_test, lr_preds))
print(confusion_matrix(y_test, lr_preds))
print(classification_report(y_test, lr_preds, target_names=le.classes_))
```

Dari Evaluasi terlihat Model **Random Forest** memberikan akurasi yang lebih tinggi, selanjutnya kita akan tingkatkan lagi agar akurasinya semakin baik dengan melakukan Hyperparameter Tuning.

### Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV

# --- 1) Definisi parameter grid ---------------------------------------
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

# --- 2) Inisialisasi dan GridSearch -----------------------------------
rf = RandomForestClassifier(random_state=42)
grid_search = GridSearchCV(rf, param_grid, cv=3, scoring='accuracy', n_jobs=-1)

# --- 3) Fit GridSearch ------------------------------------------------
grid_search.fit(X_train, y_train)

# --- 4) Evaluasi hasil terbaik ----------------------------------------
best_rf = grid_search.best_estimator_
best_preds = best_rf.predict(X_test)

print("=== Tuned Random Forest ===")
print("Best Params:", grid_search.best_params_)
print("Accuracy:", accuracy_score(y_test, best_preds))
print(confusion_matrix(y_test, best_preds))
print(classification_report(y_test, best_preds, target_names=le.classes_))
```

Terlihat Akurasi sudah mulai meningkat meski tidak signifikan, dari **74%** menjadi **75%**. Karena memang ini juga dipengaruhi oleh jumlah data kita yang sedikit dan fiktif. Untuk 75% seharusnya sudah bisa digunakan untuk prediksi. Selanjutnya kita akan simpan model dan deploy ke **Streamlit**.

### Simpan Model

```python
import joblib

# Asumsikan 'best_rf' adalah model terbaik dari GridSearchCV di sel sebelumnya
# Asumsikan 'le' adalah LabelEncoder yang digunakan untuk encoding target

# Simpan model Random Forest terbaik
joblib.dump(best_rf, 'random_forest_model.pkl')

# Simpan LabelEncoder untuk inverse_transform saat prediksi
joblib.dump(le, 'label_encoder.pkl')

print("Model Random Forest dan Label Encoder berhasil disimpan.")
```

### Testing
